# E-commerce Retail Sales in the USA

The way we consume goods and services has evolved quite drastically since we've entered the "digital age". 
* How has e-commerce developed over the year?
* Is there there any pattern in the variations of e-commerce sales?
* Can we predict how this type of sales evolves?

For this project, we study the E-Commerce Retail Sales which is a measure of U.S. "sales of goods and services where the buyer places an order, or the price and terms of the sale are negotiated over an Internet, mobile device (M-commerce), extranet, Electronic Data Interchange (EDI) network, electronic mail, or other comparable online system. Payment may or may not be made online."

The dataset is in millions of dollars, quarterly, not seasonally adjusted. The values are provided from Q4 1999 to Q1 2020.


Source: https://fred.stlouisfed.org/series/ECOMNSA

#### Test for stationarity

```python
# let's define a function to apply the Augmented Dickey-Fueller test to test data stationarity
def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df['ECOMNSA'], "E-commerce sales")
```

```python
adf_test(df['ECOMNSA'].diff(4).diff().dropna(), "Seasonally and first order differenced E-commerce sales")
```

```python
adf_test(df['log10-ECOMNSA'].diff().dropna())
```

## Modeling

### Split into train / test datasets

```python
len(df)
```

### SARIMA models

Because of the seasonality observed in the data, we'll try to fit the data with a SARIMA model. We'll work on both initial Y values and log10 transformed Y values.

#### SARIMA model

```python
auto_arima(df['ECOMNSA'], seasonal=True, m=4,
                           error_action='ignore',   # we don't want to know if an order does not work
                            suppress_warnings=True,  # we don't want convergence warnings
            ).summary()
```

```python
model = SARIMAX(train['ECOMNSA'], order = (3, 1, 3), seasonal_order = (0, 1, 1, 4)
                , enforce_invertibility= False)
```

```python
results = model.fit()
results.summary()
```

```python
start = len(train)
end = len(train) + len(test) - 1
```

```python
predictions = results.predict(start = start, end = end, dynamic = False, type = 'lev').rename('SARIMA(3, 1, 3)x(0, 1, 1, 4) predictions')
```

```python
title = 'E-Commerce Retail Sales: Real values vs predicted'
ylabel = 'Millions of Dollars, Not Seasonally Adjusted'
xlabel = ''


ax = test['ECOMNSA'].plot(figsize = (16,5), title = title, legend = True)
ax.autoscale(axis = 'x', tight = True)
ax.set(xlabel=xlabel, ylabel=ylabel)
predictions.plot(legend = True);
```

#### SARIMA model on log-Y

```python
auto_arima(df['log10-ECOMNSA'], start_p=0, start_q=0, max_p=6, max_q=6,
                               seasonal=True, m=4,
                          start_P=0, max_P=6, start_Q=0, max_Q=6, 
                          d=None, D=1, trace=True,
                          error_action='ignore',   # we don't want to know if an order does not work
                          suppress_warnings=True,  # we don't want convergence warnings
                          stepwise=True
            ).summary()
```

```python
model_log = SARIMAX(train['log10-ECOMNSA'], order = (3, 1, 0),
                   seasonal_order = (1, 1, 0, 4), enforce_invertibility= False)
```

```python
results_log = model_log.fit()
results_log.summary()
```

```python
start = len(train)
end = len(train) + len(test) - 1
```

```python
predictions_log = results_log.predict(start = start, end = end, dynamic = False,
    type = 'lev').rename('Log-Y SARIMA(3, 1, 0)x(1, 1, 0, 4)  predictions')
```

```python
title = 'E-Commerce Retail Sales: Real values vs predicted'
ylabel = 'Log10 of millions of Dollars, Not Seasonally Adjusted'
xlabel = ''


ax = test['log10-ECOMNSA'].plot(figsize = (16,5), title = title, legend = True)
ax.autoscale(axis = 'x', tight = True)
ax.set(xlabel=xlabel, ylabel=ylabel)
predictions_log.plot(legend = True);
```

Let's have a look at the results of the "Log-Y SARIMA(0, 1, 0)x(0, 1, 0, 4)" model on the original scale.

```python
predictions2 =  10 ** predictions_log
predictions2.rename('Log-Y SARIMA(3, 1, 0)x(1, 1, 0, 4) predictions', inplace = True)
```

```python
predictions.head()
```

```python
title = 'E-Commerce Retail Sales: Real values vs predicted'
ylabel = 'Millions of Dollars, Not Seasonally Adjusted'
xlabel = ''


ax = test['ECOMNSA'].plot(figsize = (16,5), title = title, legend = True)
ax.autoscale(axis = 'x', tight = True)
ax.set(xlabel=xlabel, ylabel=ylabel)
predictions.plot(legend = True);
predictions2.plot(legend = True);
```

The Log-Y SARIMA(0, 1, 0)x(0, 1, 0, 4) model seems to be a better fit. Let's check using evaluation metrics.

#### Evaluate models

```python
error1 = mean_absolute_error(test['ECOMNSA'], predictions)

error2 = np.sqrt(mean_squared_error(test['ECOMNSA'], predictions))

print(f'SARIMA(3, 1, 3)x(0, 1, 1, 4) - MAE: {error1:11.10}')
print(f'SARIMA(3, 1, 3)x(0, 1, 1, 4) - RMSE: {error2:11.10}')
```

```python
error3 = mean_absolute_error(test['ECOMNSA'], predictions2)

error4 = np.sqrt(mean_squared_error(test['ECOMNSA'], predictions2))

print(f'Log-Y SARIMA(3, 1, 0)x(1, 1, 0, 4) - MAE: {error3:11.10}')
print(f'Log-Y SARIMA(3, 1, 0)x(1, 1, 0, 4) - RMSE: {error4:11.10}')
```

As expected, the SARIMA model used on the transformed Y values provide better results. Now let's check if a Holt winter model would perform better than this model.

### Holt Winters models

#### Holt Winters model

Since our data has trend and seasonality, we build a triple exponential smoothing model. As trend is exponential and seasonality variations seem to increase in magnitude with increasing levels, a multiplicative model for both trend and seasonality might work better.

```python
fittedmodel_TESmul  = ExponentialSmoothing(train['ECOMNSA'],
                                           trend = 'mul',seasonal='mul', seasonal_periods=4).fit()
```

```python
predictions3 = fittedmodel_TESmul.forecast(17).rename('TESmul4 predictions')
```

```python
predictions3.head()
```

```python
fittedmodel_TESmix  = ExponentialSmoothing(train['ECOMNSA'],
                                           trend = 'mul',seasonal='add', seasonal_periods=4).fit()
```

```python
predictions4 = fittedmodel_TESmix.forecast(17).rename('TESmix4 predictions')
```

```python
predictions4.head()
```

```python
title = 'E-Commerce Retail Sales: Real values vs predicted'
ylabel = 'Millions of Dollars, Not Seasonally Adjusted'
xlabel = ''


ax = test['ECOMNSA'].plot(figsize = (16,5), title = title, legend = True)
ax.autoscale(axis = 'x', tight = True)
ax.set(xlabel=xlabel, ylabel=ylabel)
predictions3.plot(legend = True);
predictions4.plot(legend = True);
```

Mean absolute error and root mean squared error are lower for the multiplicative Triple Exponential Smoothing model than all the other models tested so far. Let's try to build a similar model on our log transformed dependent variable.

### Conclusion

This forecast might be used to buid a Business as usual scenario. 

We've suspected that the 2008 crisis has impacted e-commerce sales and explains the change in pattern in 2008- beginning of 2009.
We suspect that the COVID-19 pandemic is also impacting e-commerce sales on the short term:
* as people cannot physically go to most stores leading to increasing sales online,
* and e-commerce purchases might be lower, as revenue levels drop for part of the population.

On a longer term, this coronavirus crisis might drastically change consumption patterns. We might observe in the future a steeper increase of the e-commerce share in the total retail sales, as customers might prefer sales initiated online. We might also observe different trend and seasonal patterns after 2020.

When new values will be available in 2020, it would be interesting to check them against the forecast. We'll maybe also investigate the values of Advance Real Retail and Food Services Sales, especially since there're provided on a monthly basis.

By transforming our dependent variable using log10 function, we were able to get a better model performance. Our Triple Exponential Smoothing model is performing better than our best SARIMA model.
Further investigation could be building a SARIMAX model to account for exogenous data, using an indicator (dummy variable) for the period of impact of the 2008 financial crisis and check if this improves its performance level.
